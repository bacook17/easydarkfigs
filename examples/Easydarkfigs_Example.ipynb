```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import easydarkfigs
%matplotlib inline
```

The examples below show you how to use the `easydarkfigs` magic calls

```python
x = np.linspace(0, 10, 500)
y = np.cos(x)
z = np.cos(x+1.)
zerr = np.random.uniform(low=0, high=0.5, size=(len(z)))
```

## Light and Dark Styles

Let's also check we're happy with the default light and dark styles

```python
%get_lightstyle
```

```python
print(plt.style.library['easy-light'])
```

```python
with plt.style.context(('easy-light')):
    plt.plot([0, 1], label='Line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Title')
    plt.legend(loc=0)
```

```python
%get_darkstyle
```

```python
print(plt.style.library['easy-dark'])
```

```python
with plt.style.context(('easy-dark')):
    plt.plot([0, 1], label='Line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Title')
    plt.legend(loc=0)
```

If you want to change the styles that will be used to plot the light/dark figures:

```python
%set_lightstyle other-dark-style
```

# Usage

You've just spent all day making this gorgeous plot in your Jupyter notebook. Often we want to save it in two forms:

* Light-background version (for papers)

* Dark-background version (for presentations)

To do so, just wrap your Jupyter cell in the included iPython magic function `savefig_dark_light`, and give it a file name

```python
%%savefig_dark_light test_fig.jpg
fig, ax = plt.subplots()
ax.plot(x, y)
ax.errorbar(x, y=z, yerr=zerr)
ax.set_xlabel('Time')
ax.set_ylabel('Awesomeness');
```
